```python
def contains_korean(text):
    for char in text:
        if ord(char) >= 0xAC00 and ord(char) <= 0xD7A3:
            return True
    return False
```

```python
import re


def refine_url(url):

    if '.' not in url:
        return ''
    # Trim invalid leading characters (keep only letters, numbers, or valid schemes)
    trimmed_url = re.sub(r"^[^a-zA-Z0-9]+", "", url)
    
    # Trim invalid trailing characters (must end with a letter, digit, `/`, `?`, `#`, or `=`)
    trimmed_url = re.sub(r"[^a-zA-Z0-9/?#=]+$", "", trimmed_url)

    # # Ensure URL starts correctly (with a scheme or valid domain character)
    # if not re.match(r"^(https?://|ftp://|[a-zA-Z0-9])", trimmed_url):
    #     return ""  # Invalid start, return empty

    return trimmed_url

def further_filter(urll):
    url = list(urll)
    if len(url)<1:
        return ''
    chk = set(',.><-*')
    while url and url[-1] in chk:
        url.pop()
        if len(url)==0:
            break

    if 'http' in urll:
        return ''.join(url[urll.index('http'):])
    if 'www' in urll:
        return ''.join(url[urll.index('www'):])
    else:
        return ''.join(url)

    return urll


def find_first_slash_preceded_by_number(s):
    # Regular expression to find the first instance of a number followed by '/'
    match = re.search(r'\d+/', s)
    
    if match:
        return match.start() + len(match.group()) - 1  # Return the index of '/'
    else:
        return -1  # Return -1 if no match is found

def enron_refine(i):
    priority_tld = ['com','edu','org','net','biz','info','co.uk','co.nz','gov']
    other_tld = ['ru','su','uk','us','ws','ie']
    first = -1
    first = str(i).find('http')
    if first==-1:
        first = str(i).find('www.')
    last = -1
    for j in priority_tld:
        if '.'+str(j) in str(i):
            last = str(i).find(str(j))+len(j)
            break
    if last!=-1:
        return i[first:last]
    else:
        xxx = find_first_slash_preceded_by_number(i)
        if xxx!=-1:
            return i[first:xxx]
    return i
```

```python
def phone_number_count(s,m):
    filename = s
    phone_numbers = []
    has_phn_number = 0
    for i in m:
        token = str(i).split()
        for j in token:
            if j.isnumeric() and len(j)>14:
                continue
            if j.isnumeric() and len(str(j))>7:
                has_phn_number += 1
                phone_numbers.append(j)
                # print(j)
            elif j[0]=='+' and j[1:].isnumeric() and len(str(j))>7:
                has_phn_number += 1
                phone_numbers.append(j)
                # print(j)
    
    Phone_list = pd.DataFrame()
    Phone_list['Phone Number'] = phone_numbers
    # Phone_list.to_csv('../Dataset/Phone Data/'+filename+'.csv', index=None)
    print(s,has_phn_number,round(100*has_phn_number/len(m),2)) 
```

```python
import re

def email_count(s,m):
    filename = s
    emails = []
    has_email = 0
    for i in m:
        token = str(i).split()
        for idx,j in enumerate(token):
            if ('@' in j and '.' in j and '/' not in j and (j[0]!='@' or j[-1]!='@')) :
                has_email += 1
                if len(j)<7:
                    emails.append(''.join(token[idx-7:idx-1])+j+''.join(token[idx+1:idx+8]))
                
                else:
                    emails.append(j)
                
                break
                # print(j)

    Email_list = pd.DataFrame()
    Email_list['Email'] = emails
    # Email_list.to_csv('../Dataset/Email Data/'+filename+'.csv', index=None)
    print(s,has_email,round(100*has_email/len(m),2)) 
```
